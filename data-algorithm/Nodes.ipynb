##  _Node_
+ _Making Node for making link between each others eg. In [3]_



```python
class Node: #This is a Node 
    def __init__(self,data=None):
        self.data= data   # make object for data (works as data part)
        self.next= None   # make object points none (works as pointer)
        self.back = None  #same as next object
        
    def __str__(self):     #string format (not important)
        return str(data)
```

```python
a=Node('1');print(a.data)  #object assign with data
b=Node('2');print(b.data)  #object assigned with data 
```

```python
a.next=b  #make link between node a to b 
b.back=a  #make link between node b to a
# example
```

```python
start = a
end = b
while start:
    print(start.data)
    start=start.next
print('Here starts the second pointer...')
while end:
    print(end.data)
    end=end.back        
```
